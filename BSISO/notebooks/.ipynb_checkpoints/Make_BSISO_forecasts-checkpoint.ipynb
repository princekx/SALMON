```python
def running_mean(x, N, axis=0):
    cumsum = np.cumsum(np.insert(x, 0, 0, axis=0), axis=0) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)
```

```python
def return_indices_of_adf(adf, bdf):
    '''
    Returns the indices of DataFrame A that matches elements in DataFrame B

    :param adf:
    :type adf:
    :param bdf:
    :type bdf:
    :return:
    :rtype:
    '''
    a = [year * 10000 + month * 100 + day for year, month, day in zip(adf.year, adf.month, adf.day)]
    b = [year * 10000 + month * 100 + day for year, month, day in zip(bdf.year, bdf.month, bdf.day)]
    b_set = set(b)
    return [int(i) for i, v in enumerate(a) if v in b_set]
```

```python
def make_composite(cube, inds, lag=20):
    # Get a mid value of the number of indices
    if len(inds)>3:
        n2 = 3
    comp_cube = cube[inds[n2]-lag:inds[n2]+lag+1].copy()
    comp_data = []
    for ind in inds:
        #print(ind)
        try:
            comp_data.append(cube[ind-lag:ind+lag+1].data)
        except:
            print('No date found. Skip')
            pass        
    comp_cube.data = np.ma.mean(np.array(comp_data), axis=0)
    return comp_cube
```

```python
def makecube(data_array, lats=None, lons=None, varname='OLR_1'):
    """
    Create an Iris cube from a 2D NumPy array with latitude and longitude coordinates.

    Args:
        data_array (numpy.ndarray): 2D NumPy array containing the data values.
        lats (numpy.ndarray or None): 1D NumPy array of latitude values (optional).
        lons (numpy.ndarray or None): 1D NumPy array of longitude values (optional).
        varname (str): Variable name for the cube (default: 'OLR_1').

    Returns:
        iris.cube.Cube: Iris cube object containing the data and coordinates.
    """
    # Create latitude and longitude coordinates if not provided
    if lats is None:
        lats = iris.coords.DimCoord(range(data_array.shape[0]), standard_name='latitude', units='degrees_north')
    else:
        lats = iris.coords.DimCoord(lats, standard_name='latitude', units='degrees_north')
    if lons is None:
        lons = iris.coords.DimCoord(range(data_array.shape[1]), standard_name='longitude', units='degrees_east')
    else:
        lons = iris.coords.DimCoord(lons, standard_name='longitude', units='degrees_east')

    # Ensure the latitude and longitude arrays are 1D
    if lats.ndim != 1:
        raise ValueError("Latitude array must be 1-dimensional.")
    if lons.ndim != 1:
        raise ValueError("Longitude array must be 1-dimensional.")

    # Create cube from the data array
    cube = iris.cube.Cube(data_array, var_name=varname)

    # Add latitude and longitude coordinates to the cube
    cube.add_dim_coord(lats, 0)
    cube.add_dim_coord(lons, 1)

    return cube
```

```python
def prepare_calendar(cube):
    # Setting up the dates on data
    for coord_name, coord_func in [('year', iris.coord_categorisation.add_year),
                                   ('month_number', iris.coord_categorisation.add_month_number),
                                   ('day_of_month', iris.coord_categorisation.add_day_of_month),
                                   ('day_of_year', iris.coord_categorisation.add_day_of_year),
                                   ('hour', iris.coord_categorisation.add_hour)]:
        if not cube.coords(coord_name):
            coord_func(cube, 'forecast_time', name=coord_name)
    return cube

def create_dates_df(cube):
    cube = prepare_calendar(cube)
    cube_dates_dt = [datetime.datetime(y, m, d) for y, m, d in zip(cube.coord('year').points,
                                                                   cube.coord('month_number').points,
                                                                   cube.coord('day_of_month').points)]
    # Extract year, month, and day from each datetime object
    data = {
    'Year': [dt.year for dt in cube_dates_dt],
    'Month': [dt.month for dt in cube_dates_dt],
    'Day': [dt.day for dt in cube_dates_dt]
    }
    
    # Create a Pandas DataFrame
    df = pd.DataFrame(data)
    
    # Create a new 'date' column with datetime objects
    df['date'] = pd.to_datetime(df[['Year', 'Month', 'Day']])
    return df
```

```python
def calculate_mjo_phase(rmm1, rmm2):
    # Calculate the phase angle
    phase_angle = np.arctan2(rmm2, rmm1)
    
    # Convert negative angles to positive
    phase_angle = np.where(phase_angle < 0, phase_angle + 2 * np.pi, phase_angle)

    # Define the phase categories based on the phase angle
    phase = np.zeros(len(phase_angle), dtype=int)
    phase = np.where((phase_angle >= 0) & (phase_angle < np.pi / 4), 5,
                     np.where((phase_angle >= np.pi / 4) & (phase_angle < np.pi / 2), 6,
                              np.where((phase_angle >= np.pi / 2) & (phase_angle < 3 * np.pi / 4), 7,
                                       np.where((phase_angle >= 3 * np.pi / 4) & (phase_angle < np.pi), 8,
                                                np.where((phase_angle >= np.pi) & (phase_angle < 5 * np.pi / 4), 1,
                                                         np.where((phase_angle >= 5 * np.pi / 4) & (phase_angle < 3 * np.pi / 2), 2,
                                                                  np.where((phase_angle >= 3 * np.pi / 2) & (phase_angle < 7 * np.pi / 4), 3, 4)))))))

    return phase
```

```python
def read_bsiso_eofs():
    eof_data_dir = '../data'
    eof_data_file = os.path.join(eof_data_dir, 'BSISO.EOFstruc.data')
    eofs = pd.read_csv(eof_data_file, delim_whitespace=True)

    nlat = eofs['YY'].max()
    nlon = eofs['XX'].max()
    lons = np.linspace(40, 160, nlon)
    lats = np.linspace(-10.,40., nlat)

    comb_eof1 = np.concatenate((eofs.OLR1, eofs.U8501), axis=0)
    comb_eof2 = np.concatenate((eofs.OLR2, eofs.U8502), axis=0)

    return comb_eof1, comb_eof2    
```

```python
def compute_bsiso_indices(date, mem, nanalysis2write=40, model='mogreps'):
    date_label = date.strftime("%Y%m%d")
    # Read data
    olr_cube = iris.load_cube(f'/scratch/hadpx/SEA_monitoring/processed_SEA_data/{model}/MJO/olr/olr_120dm_40sn_nrt_{date_label}_{mem}.nc')
    u850_cube = iris.load_cube(f'/scratch/hadpx/SEA_monitoring/processed_SEA_data/{model}/MJO/u850/u850_120dm_40sn_nrt_{date_label}_{mem}.nc')
    
    reg_constraints = iris.Constraint(latitude=lambda cell: -10 <= cell <= 40) & iris.Constraint(longitude=lambda cell: 40. <= cell <= 160.)
    
    olr_cube = olr_cube.extract(constraint=reg_constraints)
    u850_cube = u850_cube.extract(constraint=reg_constraints)

    # regional standard deviation of analysis
    olr_std = olr_cube[:nanalysis2write].collapsed(('forecast_time', 'latitude', 'longitude'), iris.analysis.STD_DEV)
    u850_std = u850_cube[:nanalysis2write].collapsed(('forecast_time', 'latitude', 'longitude'), iris.analysis.STD_DEV)

    print(olr_std.data, u850_std.data)
    # Create pcs 
    pcs = create_dates_df(olr_cube)

    # Doing running mean
    N = 7 # odd number
    #olr_cube.data[int(N/2):-int(N/2)] = running_mean(olr_cube.data, N)
    #u850_cube.data[int(N/2):-int(N/2)] = running_mean(u850_cube.data, N) 
    
    ntime, nlat_cube, nlon_cube = olr_cube.shape
    print(ntime, nlat_cube, nlon_cube)
    #if (nlat_cube == nlat) & (nlon_cube==nlon):
    #    print('Cube dims match')
    olr_resh = np.reshape(olr_cube.data, (ntime, -1))
    u850_resh = np.reshape(u850_cube.data, (ntime, -1))
    
    #normalise
    #olr_resh = (olr_resh - olr_resh.mean())/olr_resh.std()
    #u850_resh = (u850_resh - u850_resh.mean())/u850_resh.std()
    olr_resh = olr_resh/olr_std.data
    u850_resh = u850_resh/u850_std.data

    
    # Combining the data along the lat-lon dimensions
    comb_fcast_data = np.concatenate((olr_resh, u850_resh), axis=1)

    # Read Predefined EOF structures
    comb_eof1, comb_eof2 = read_bsiso_eofs()
    
    # Assuming comb_eof1 is a 1D array of shape (2058,)
    # and comb_fcast_data is a 2D array of shape (261, 2058)
    pcs['PC1'] = np.dot(comb_fcast_data, comb_eof1)
    pcs['PC2'] = np.dot(comb_fcast_data, comb_eof2)
    
    # Columns to normalize
    columns_to_normalize = ['PC1', 'PC2']
    
    # Normalization by mean and standard deviation for specific columns
    for col in columns_to_normalize:
        pcs[col][:nanalysis2write] = (pcs[col][:nanalysis2write] - pcs[col][:nanalysis2write].mean()) / pcs[col][:nanalysis2write].std()

    # Normalization by mean and standard deviation for specific columns
    for col in columns_to_normalize:
        pcs[col][nanalysis2write:] = (pcs[col][nanalysis2write:] - pcs[col][nanalysis2write:].mean()) / pcs[col][nanalysis2write:].std()
    
    # calculate phase of just the forecasts
    #pcs = pcs[-61:]
    
    #df["day"] = pd.to_datetime(df["year"] * 1000 + df["julian day"]).dt.day
    #pcs[pcs<-900] = 0 #np.NaN
    
    pcs['8_phases'] = calculate_mjo_phase(pcs['PC1'].values, pcs['PC2'].values)
    pcs['amp'] = np.sqrt(pcs['PC1'].values**2+pcs['PC2'].values**2)
    pcs['mem'] = [mem for i in range(len(pcs))]
    
    # Create the 'Lee_phase' column based on the condition
    pcs['4_phases'] = 0  # Initialize the column with 0
    pcs.loc[pcs['8_phases'].isin([8, 1]), '4_phases'] = 1
    pcs.loc[pcs['8_phases'].isin([2, 3]), '4_phases'] = 2
    pcs.loc[pcs['8_phases'].isin([4, 5]), '4_phases'] = 3
    pcs.loc[pcs['8_phases'].isin([6, 7]), '4_phases'] = 4

    return pcs    
```

```python
def plot_image_map(plot, cube, **kwargs):
    palette = kwargs['palette']
    if 'cbar_title' in kwargs and kwargs['cbar_title']:
        cbar_title = kwargs['cbar_title']
    else:
        cbar_title = None
    if 'palette_reverse' in kwargs and kwargs['palette_reverse']:
        palette = palette[::-1]
    lons = cube.coord('longitude').points
    lats = cube.coord('latitude').points

    color_mapper = LinearColorMapper(palette=palette, low=kwargs['low'], high=kwargs['high'])


    plot.image(image=[cube.data], x=min(lons), y=min(lats), dw=max(lons) - min(lons),
               dh=max(lats) - min(lats), color_mapper=color_mapper, alpha=0.7)


    plot.x_range = Range1d(start=min(lons), end=max(lons))
    plot.y_range = Range1d(start=min(lats), end=max(lats))

    color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12,
                         border_line_color=None, location=(0, 0),
                         orientation='horizontal', title=cbar_title)
    #plot.add_layout(color_bar, 'below')
    return plot
```

```python
def running_mean(x, N, axis=0):
    cumsum = np.cumsum(np.insert(x, 0, 0, axis=0), axis=0) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)
```
